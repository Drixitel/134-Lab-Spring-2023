```python
plt.style.use('fivethirtyeight')

mu = -1.602E-19
sigma = 1.602E-21

z = (e_val - e_measured)/sig_e
Zscore = (e_measured - mu)/sigma



cdf = sc.stats.norm.cdf(Zscore, 0,1)
print('CDF from the normalized distribution= ', cdf)
print(f"Zscore = {Zscore}")


if cdf > 0.5: 
    pvalue_2t = 2*(1-cdf)
else: 
    pvalue_2t = 2*(cdf)
print("\nTwo tailed p-value = ",pvalue_2t )
print("This represents our cut-off region,\nzscores in these region reject the null hypothosis.")



fig, ax = plt.subplots(figsize = (9,6))

plt.hist(z, 25, edgecolor ='black', color= 'navy', alpha=0.65) # log =True give more but... not needed
plt.tight_layout()
plt.xlabel("Charge Value")
plt.ylabel("Droplets")
plt.yticks(range(0,8,1))


# ax.vlines(x = Zscore, ymax= 6, ymin= 0, color ='purple', label = 'Zscore = 1.37')
plt.axvline(x = 0.167, color ='gray', label = 'T-value= $\pm$ 0.167')
plt.axvline(x = -0.167, color ='gray')
plt.axvline(x = Zscore, color ='purple', label = 'Zscore = 1.37')

# plt.vlines(x=-1.602, ymin = 0, ymax = 6, color = 'red', ls= '--', label= "e = -1.602E-19")
plt.legend()
plt.show()
```
